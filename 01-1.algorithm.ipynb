#세 정수를 입력받아 최댓값 구하기

```python
print('세 정수의 최댓값을 구합니다.')
a=int(input('정수 a의 값을 입력: '))
b=int(input('정수 b의 값을 입력: '))
c=int(input('정수 c의 값을 입력: '))

maximum=a
if b>maximum: maximum=b
if c>maximum: maximum=c

print(f'최댓값은 {maximum}입니다')
```

#세 정수의 중앙값 구하기

```python
def med(a,b,c):
    if a>=b:
        if b>=c:
            return b
        elif a>=c:
            return c
        else:
            return a
    elif b>=c:
        if a>=c:
            return a
        else: 
            return c
    else:
        return b

print(med(1,2,3))
print(med(2,3,1))
print(med(1,4,3))
print(med(1,1,5))
print(med(2,3,3))
```

#입력받은 정수의 부호(양수, 음수, 0) 출력하기  # 3개 분기문

```python
n = int(input())
if n>0:
    print("이 수는 양수이다.")
elif n<0:
    print("이 수는 음수이다.")
else:
    print("이 수는 0이다.")
```
