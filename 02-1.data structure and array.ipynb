튜플

```python
# 튜플은 원소에 순서를 매겨 결합한 것으로 원소를 변경할 수 없는 immutable 자료형이다.
# 내장함수인 tuple()을 이용하면 ()안에 문자열/리스트/집합/range()함수를 넣을 수 있음.
```

```python
# 좌변에 여러 개의 변수를 놓고 우변에 리스트나 튜플을 놓으면, 우변의 원소를 좌변의 변수에 한 번에 대입할 수 있다. 
x=[1,2,3]
a,b,c = x
a,b,c
```

인덱스로 원소에 접근하기

```python
x = [11,22,33,44,55,66,77]
x[2]
x[-3]
x[-4] = 3.14
x
```

슬라이스식으로 원소에 접근하기

```python
# s[i:j]  i번째 원소부터 j-1번째 원소까지 나열
# s[i:j:k] i번째 원소부터 j-1번째 원소까지 k씩 건너뛰며 나열
x[1:4]
x[1:6:2]
x[-3:-1]
```
