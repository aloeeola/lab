1부터 n까지 정수의 합 구하기 (while 문)

```python
print("지금부터 1부터 n까지의 합을 구하겠습니다.")
n=int(input('n값을 입력하시오: '))

sum=0
i=1

while i <= n:   #i가 n보다 작거나 같을 때까지 반복
    sum += i
    i += 1

print(f"1부터 {n}까지의 합은 {sum}입니다.")
print(f"i값은 {i}입니다.")
```

1부터 n까지 정수의 합 구하기 (for 문) 

변수가 하나만 있을 때는 while문보다 for문 사용하는 것이 더 좋다

```python
print("1부터 n까지의 합을 구하자.")
n = int(input("n의 값을 입력하시오.: "))

sum = 0 
for i in range(1,n+1):
    sum += i 

print(f"1부터 {n}까지의 합은 {sum}이다.")
print(f"i={i}")

# iterable: 반복할 수 있는 개체 i를 말함 , range()함수가 iterable object를 생성하는 것
# range(n): 0이상 n미만인 수를 차례로 나열하는 수열
# range(a,b): a이상 b미만인 수를 차례로 나열하는 수열
# range(a,b,n): a이상 b 미만인 수를 n간격으로 나열하는 수열

# PYTHON iterable object: list, str, tuple
```

a부터 b까지 정수의 합 구하기 (for문)

```python
print("a부터 b까지의 합을 구하자.")
a = int(input("a=?: "))
b = int(input("b=?: "))

if a > b:
    a,b = b,a       # a와 b의 값을 교환(단일 대입문 사용)

sum=0
for i in range(a,b+1):
    sum += i
print(f"{a}부터 {b}까지의 합은 {sum}이다.")
```

반복 과정에서 조건 판단하기 1

```python
#a부터 b까지 정수의 합을 구하는 과정과 최종값을 구하는 프로그램
print("a부터 b까지의 합을 구하자.")
a = int(input("a=?: "))
b = int(input("b=?: "))

if a > b:
    a,b = b,a       

sum=0
for i in range(a,b+1):
    if i<b: print(f"{i}+",end='')       #print('~~', end=''): ~~출력하고 그 아래 출력물을 바로 옆에 이어서 출력하기 
    else: print(f"{i}=", end='')
    sum += i

print(sum)
```

```python
print("a부터 b까지의 합을 구하자.")
a = int(input("a=?: "))
b = int(input("b=?: "))

if a > b:
    a,b = b,a 

sum=0

for i in range(a,b):
    print(f"{i}+", end='')
    sum += i
sum += b

print(f"{b}={sum}")
```

반복 과정에서 조건 판단하기 2

```python
# +와 -를 번갈아 출력하기 1 _ 나눗셈 n번 시행

print("+와 -를 번갈아 출력합니다.")
n = int(input("몇 개를 출력할까요?: "))
for i in range(n):
    if i%2==0:
        print('+',end='')
    else:
        print('-',end='')
```

```python
# +와 -를 번갈아 출력하기 2 _ 나눗셈 2번 시행

print("+와 -를 번갈아 출력합니다.")
n = int(input("몇 개를 출력할까요?: "))
for _ in range(n//2):
    print('+-',end='')
if n%2==1:
    print('+',end='')
```

양수만 입력받기

```python
while True:
    n = int(input('n값을 입력하시오.: '))
    if n>0:
        break       # n이 0보다 커질때까지 반복
sum = 0

for i in range(1,n+1):
    sum += i
    i += 1

print(sum)
```

가로,세로 길이가 정수이고 넓이가 area인 직사각형에서 변의 길이 나열하기

```python
area = int(input('직사각형의 넓이를 입력하세요: '))

for i in range(1, area+1):
    if i * i > area : break
    if area % i: continue
    print(f'{i}*{area//i}')
```

```python
#오른쪽 아래가 직각인 이등변 삼각형 모양으로 * 출력하기
n = int(input('직각이등변삼각형의 짧은 변의 길이를 입력하시오: '))
for i in range(n):
    for j in range(n-i):
        print(' ',end='')
    for _ in range(i+1):
        print('*',end='') 
    print()
```
